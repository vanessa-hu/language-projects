```python
#actual code

from time import sleep

char1 = [["昔", "former times"],
         ["xī", [["昔日","xīrì","formerly"],["昔年","xīnián","former years"]]], 
        [ "", []], 
         ["むかし (mukashi)",[["昔","むかし (mukashi)","old times"],["昔話","むかしはなし (mukashihanashi)","old tale"]] ]]

char2 = [["々", "kanji repetition"],
         ["", []], 
        [ "", []], 
         ["[sound repetition]",[["昔々","むかしむかし (mukashimukashi)","once upon a time"],["人々","ひとびと (hitobito)","people"]] ]]

char3 = [["神", "God, spirit"],
         ["shén", [["神奇","shénqí","mystical"],["神话","shénhuà","myth"],["神秘","shénmì","mysterious"]]], 
        [ "じん (jin)；しん (shin)", [ ["神話","しんわ (shinwa)","myth"], ["神秘な","しんぴ (shinpina)","mysterious"] ]], 
         ["むかし (mukashi)",[["昔","むかし (mukashi)","old times"],["昔話","むかしはなし (mukashihanashi)","old tale"]] ]]

char4 = [["早", "early"],
         ["zǎo", [["早晨","zǎochén","early morning"],["早春","zǎochūn","early spring"],["早出晚归","zǎochūwǎnguī","to leave early and return late"]]], 
        [ "そう (kou)", [ ["早朝","そうちょう (souchou)","early morning"], ["早春","そうしゅん (soushun)","early spring"] ]], 
         ["はや (haya)",[["早い","はやい (hayai)","early"],["早口","はやくち (hayakuchi)","fast-talking"]] ]]

char5 = [["起", "get up, start"],
         ["qǐ", [["起源","qǐyuán","origin"],["起诉","qǐsù","to prosecute"],["起居","qǐjū","daily activities"]]], 
        [ "き (ki)", [ ["起原","きげん (kigen)","origin"], ["起诉","きそ (kiso)","prosecution"] ]], 
         ["お (o)",[["起きる","おきる (okiru)","to get up"],["起こす","おこす (okosu)","to wake someone up"]] ]]

char6 = [["牛", "cow"],
         ["niú", [["牛奶","niúnǎi","cow's milk"],["牛气冲天","niúqichōngtiān","very arrogant"]]],
        [ "ぎゅう (gyuu)", [ ["牛乳","ぎゅうにゅう (gyuunyuu)","cow's milk"], ["牛丼","ぎゅうどん (gyuudon)","beef bowl rice dish"] ]], 
         ["うし (ushi)",[["牛","うし (ushi)","cow"],["子牛","こうし (koushi)","calf"]] ]]

char7 = [["使", "use"],
         ["shǐ", [["使尽","shǐjìn","to exert all one's strength"],["大使"," dàshǐ","ambassador"],["使用","shǐyòng","to make use of"]]],
        [ "し (shi)", [ ["大使","たいし (taishi)","ambassador"], ["使用","しよう (shiyou)","use; application"] ]], 
         ["つか (tsuka)",[["使う","つかう (tsukau)","to use"],["使い道","つかいみち (tsukaimichi)","purpose"]] ]]

char8 = [["働", "work","动"],
         ["dòng", [["劳动","láodòng","physical labor"],["动怒","dòngnù","to get angry"],["动议","dòngyì","proposal"]]], 
        [ "どう (dou)", [ ["労働","ろうどう (roudou)","labor"] ]], 
         ["はたら (hatara)",[["働く","はたらく (hataraku)","to work"],["働かせる","はたらかせる (hatarakaseru)","to apply; to exert"]] ]]

char9 = [["連", "to link","连"],
         ["lián", [["连接","liánjiē","link"],["连通","liántōng","to communicate"],["连连","liánlián","repeatedly"]]], 
        [ "れん (ren)", [ ["国連","こくれん (kokuren)","United Nations"], ["連休","れんきゅう (renkyuu)","consecutive holidays"] ]], 
         ["つ (tsu)",[["連れて帰る","つれてかえる (tsuretekaeru)","to bring (a person) back"],["連ねる","つらねる (tsuraneru)","to link; to put together"]] ]]

char10 = [["別", "other, separate"],
         ["bié", [["特别","tèbié","particular, especially"],["差别","chābié","difference"],["别扭","bièniu","awkward"]]], 
        [ "べつ (betsu)", [ ["特別な","とくべつな (tokubetsuna)","special"], ["差別","さべつ (sabetsu)","discrimination/differentiation"], ["別に","べつに (betsuni)","not in particular"]]], 
         ["わか (waka)",[["別れる","わかれる (wakareru)","to break up"]] ]]


char11 = [["度", "time; degrees"],
         ["dù", [["态度","tàidu","attitude"],["一度","yīdù","at one time"],["温度","wēndù","temperature"]]], 
        [ "ど (do)", [ ["態度","たいど (taido)","attitude"], ["一度","いちど (ichido)","once"], ["温度","おんど (ondo)","temperature"] ]], 
         ["",[]] ]

char12 = [["赤", "red, bare"],
         ["chì", [["赤道","chìdào","equator"],["赤诚","chìchéng","wholly devoted"],["赤子","chìzǐ","subjects; newborn baby"]]], 
        [ "せき (seki)", [ ["赤道","せきどう (seidou)","equator"], ["赤面","せきめん (sekimen)","blushing"] ]], 
         ["あか (aka)",[["赤い","あかい (akai)","red"],["赤ちゃん","あかちゃん (akachan)","baby"]] ]]

char13 = [["青", "blue, green"],
         ["qīng", [["青少年","qīngshàonián","young person"],["青春","qīngchūn","youth; springtime of life"],["青翠","qīngcuì","verdant"]]], 
        [ "せい (sei)", [ ["青少年","せいしょうねん (seishounen)","young person"], ["青春","せいしゅん (seishun)","youth; springtime of life"] ]], 
         ["あお (ao)",[["青い","あおい (aoi)","blue"],["青信号","あおしんごう (aoshingou)","green light"]] ]]

char14 = [["色", "color"],
         ["sè", [["特色","tèsè","characteristic"],["景色","jǐngsè","scenery"],["色调","sèdiào","hue"]]], 
        [ "しき (shiki)、しょく (shoku)", [ ["特色","とくしょく (tokushoku)","characteristic"], ["景色","けしき (keshiki)","scenery"] ]], 
         ["いろ (iro)",[["色","いろ (iro)","color"],["色々な","はやくち (いろいろな (iroirona)","various"]] ]]


kanjiList = {'昔': char1, '々': char2, '神': char3, '早': char4, '起': char5, '牛': char6, '使': char7,
             '働': char8, '連': char9, '別': char10, '度': char11, '赤': char12,'青': char13, '色': char14 }
             
currentKanji = ""
currentList = []
currentSearch = ""


def greeting():
    print("Hello! I am the the Genki Lesson 12 Kanji Chatbot. Let's learn kanji together!\n")
    sleep(2)
    print("Let's get started.  Lesson 12 has this list of 12 kanji: 昔, 々, 神, 早, 起, 牛, 使, 働, 連, " 
          +"別, 度, 赤, 青, 色.\n")
    sleep(3)
    print("When prompted, type the number of the choice you want to go with.\n")
    sleep(1)
    print("[1]: Kanji meaning.\n")
    sleep(1)
    print("[2]: Chinese info.\n")
    sleep(1)
    print("[3]: On-yomi info. \n")
    sleep(1)
    print("[4]: Kun-yomi info. \n")
    sleep(1)
    print("[5]: Japanese and Chinese similarities. \n")
    sleep(1)
    print("[6]: Examples with this kanji.\n")
    sleep(1)
    print("[7]: Switch to another kanji.\n")
    sleep(1)
    print("[8]: Goodbye!\n")
    sleep(1)
    
def newSearch():
    global currentKanji
    global currentList
    global currentSearch
    currentKanji = ""
    currentList = []
    currentSearch = ""
    print("\nRemember, the list of possible kanji is 昔, 々, 神, 早, 起, 牛, 使, 働, 連, " 
          +"別, 度, 赤, 青, 色.")
    currentKanji = input("\nNow type a kanji character from the given list to find out more about it! ")
    sleep(2)
    if currentKanji in kanjiList:
            currentList = kanjiList[currentKanji]           
            
            print("\nOkay, awesome! What do you want to know about \"" + currentKanji + "\"? ")
            showOptions()
             
    else:
        print ("\nSorry, I don't understand. Please type a valid kanji. ")
        newSearch()
        
        
        
        
def kanjiMeaning():
    global currentList
    global currentKanji
    sleep(2)
    print("\nThe character {kanji} means {meaning}.".format(kanji = currentKanji,meaning = currentList[0][1]))
    sleep(2)
    print("\nWhere do you want to go from here?")
    showOptions()

def compare():
    global currentKanji
    global currentList
    sleep(2)
    if currentList[2][0] == "":
        print("\nSorry, there is no Japanese pronunciation derived from the original Chinese character!")
    else:
        print(("\nThe on-yomi pronunciation of " + currentKanji + 
               " is {on}, which is derived from Chinese pronunciation {ch}.").format(on = currentList[2][0], ch = currentList[1][0]))
        sleep(2)
        if len(currentList[0]) == 3:
            print("\nIn Chinese, a slightly different character -- {ch} -- is used instead of {jp}.".format(ch = currentList[0][2],jp=currentKanji))
            sleep(2)
        print("\nNote the similarities in the examples below.")
        sleep(2)
        examples(1)
        examples(2)
    print("\nThat was interesting, right? What else do you want to know?")
    showOptions()
    
    
def typeInfo(type):
    #parameter = 1; Chinese. = 2; on-yomi. = 3; kun-yomi.
    global currentList
    global currentKanji
    name = ""
    char = currentKanji
    if type ==1:
        name = "Chinese"
    elif type ==2:
        name = "on-yomi"
    elif type ==3:
        name = "kun-yomi"
    if currentList[type][0] == "":
        print("\nThis kanji doesn't have a " + name + " pronunciation.")
    else:
        if len(currentList[0]) == 3 and type == 1:
            char = currentList[0][2]
            print("\nIn Chinese, a slightly different character -- {ch} -- is used instead of {jp}.".format(ch = char,jp=currentKanji))
            
        print("\nThe " + name + " pronunciation of {kanji} is {pro}.".format(kanji = char,pro = currentList[type][0]))
        sleep(2)
        examples(type)
       
    sleep(2)
    print("\nGreat, what else would you like?")
    showOptions()
    
        
def examples(type):
    name = ""
    if type ==1:
        name = "Chinese"
    elif type ==2:
        name = "On-yomi"
    elif type ==3:
        name = "Kun-yomi"

    print("\n" + name + " examples (if any) include: ")
    for list in currentList[type][1]:
        if currentList[type][1] != []:
            sleep(2)
            print("\n{phrase}, pronounced {pro}, which means {meaning}.".format(phrase = list[0], pro = list[1], meaning = list[2]))
    sleep(2)

    
def goodbye():
    print("\nThanks for studying! See you next time.")
    global currentKanji
    global currentList
    global currentSearch
    currentSearch = ""
    currentList = []
    currentKanji = ""
    
def showOptions():
    global currentSearch
    sleep(2)
    currentSearch = input("\nType the number of the choice you select: ")

   
    if currentSearch == "1":
        kanjiMeaning()
    elif currentSearch == "2":
        typeInfo(1)
    elif currentSearch == "3":
        typeInfo(2)
    elif currentSearch == "4":
        typeInfo(3)
    elif currentSearch == "5":
        compare()
    elif currentSearch == "6":
        examples(1)
        examples(2)
        examples(3)
        print("\nOkay, now what would you like to do?")
        showOptions()
    elif currentSearch == "7":
        newSearch()
    elif currentSearch == "8":
        goodbye()
    else:
        print("\nSorry, we didn't understand you! Please try again.")
        showOptions()

greeting()
newSearch()


```
